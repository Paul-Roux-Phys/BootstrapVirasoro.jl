```julia
using BootstrapVirasoro
using SpecialFunctions
using BootstrapVirasoro.LoopModels
setprecision(BigFloat, 40, base=10);
```

```julia
β = big"0.9" + big"0.6"*im;
print(β^2)
c = CC(β = β);
```

```julia
s0 = 1;

P1 = 0.67;
P2 = 0.98 - 0.23im;

D1 = CD(c,P=P1);
D2 = CD(c,P=P2);

Cor = Correlation(D1,D2,D2,D1,30);

P = 0.23;
VPk = [CD(c,P=P+k/β) for k in -10:10];

Brs0 = [CD(c,r=R,s=s0) for R in 1:2:11];

fs(V) = CBlock(Cor,:s,V);
ft(V) = CBlock(Cor,:t,V);

Cs = ChannelSpectrum(Cor,:s,VPk,fs);
Ct = ChannelSpectrum(Cor,:t,Brs0,ft);
```

```julia
sol = solve_bootstrap(Channels(Dict(:s => Cs,:t => Ct)));
```

```julia
specs = Channels(Dict(:s => Cs,:t => Ct))
```

```julia
sys = BootstrapSystem(specs)
```

```julia
evaluate_blocks!(sys)
```

```julia
z = 0.5 + 0.3im
println(BootstrapVirasoro.evaluate_correlation(sys, :s, z))
```
